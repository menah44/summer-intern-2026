## 1. Git vs GitHub — What's the Difference?

| | **Git** | **GitHub** |
|---|---------|-----------|
| **What is it?** | A **tool** installed on your computer | A **website** (github.com) |
| **Purpose** | Track changes in your code (version control) | Store & share your code online |
| **Works offline?** | ✅ Yes — runs locally | ❌ No — needs internet |
| **Created by** | Linus Torvalds (2005) | Microsoft / GitHub Inc. (2008) |
| **Example** | `git commit -m "fix bug"` | Pull Requests, Issues, Forks |

### Simple analogy:
- **Git** = a diary where you write every change you make
- **GitHub** = a cloud library where you store and share that diary with others

### Git Workflow (local)
```
git init                          # Start tracking a project
git add .                         # Stage all changes
git commit -m "my first commit"   # Save a snapshot
git log --oneline                 # View history
```

### GitHub Workflow (remote)
```
git remote add origin <url>       # Connect to GitHub repo
git push origin main              # Upload your code
git pull origin main              # Download latest changes
git clone <url>                   # Copy a repo from GitHub
```

> **Key point:** You use **Git** commands on your computer, and **GitHub** is where the code lives online so your team can collaborate.

## 3. NumPy — Numerical Computing

**NumPy** lets you work with arrays of numbers very fast (much faster than Python lists).

**When to use it:** math operations, matrices, statistics.

### Create Arrays

<div style="background-color: #f0fdf4; padding: 15px; border-left: 5px solid #22c55e; border-radius: 5px; margin-bottom: 10px;">
    <h4 style="color: #166534; margin-top: 0;">🎯 What are we doing here?</h4>
    <p style="color: #15803d; margin-bottom: 0; font-size: 14px;">Creating 1D and 2D arrays (matrices) using NumPy to understand basic array structures and shapes.</p>
</div>

````python
import numpy as np

a = np.array([1, 2, 3, 4, 5])         # 1D array
print("Array:", a)

matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])          # 2D array
print("Matrix:\n", matrix)
print("Shape:", matrix.shape)

# Quick ways to create arrays
print("Zeros:", np.zeros(4))
print("Range:", np.arange(0, 10, 2))
````

### Math Operations (no loops needed!)

<div style="background-color: #f0fdf4; padding: 15px; border-left: 5px solid #22c55e; border-radius: 5px; margin-bottom: 10px;">
    <h4 style="color: #166534; margin-top: 0;">🎯 What are we doing here?</h4>
    <p style="color: #15803d; margin-bottom: 0; font-size: 14px;">Performing basic mathematical operations and calculating core statistics (mean, max, sum) on NumPy arrays.</p>
</div>

````python
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

print("a + b =", a + b)
print("a * 2 =", a * 2)
print("Mean  =", np.mean(a))
print("Max   =", np.max(a))
print("Sum   =", np.sum(a))
````

### Handle Missing Data

<div style="background-color: #f0fdf4; padding: 15px; border-left: 5px solid #22c55e; border-radius: 5px; margin-bottom: 10px;">
    <h4 style="color: #166534; margin-top: 0;">🎯 What are we doing here?</h4>
    <p style="color: #15803d; margin-bottom: 0; font-size: 14px;">Simulating missing data and demonstrating how to handle it by filling in the blanks with the average values.</p>
</div>

````python
# Simulate missing values
df_test = df[["name", "gpa"]].copy()
df_test.loc[0, "gpa"] = None
df_test.loc[3, "gpa"] = None

print("Missing values:", df_test["gpa"].isnull().sum())
print(df_test.head())
#Fill with the average
df_test["gpa"] = df_test["gpa"].fillna(df_test["gpa"].mean())
print("After filling:", df_test["gpa"].isnull().sum())
````

## 5. Matplotlib — Data Visualization

**Matplotlib** creates charts and graphs from your data.

**When to use it:** showing trends, comparing groups, presenting results.

### Bar Chart

<div style="background-color: #f0fdf4; padding: 15px; border-left: 5px solid #22c55e; border-radius: 5px; margin-bottom: 10px;">
    <h4 style="color: #166534; margin-top: 0;">🎯 What are we doing here?</h4>
    <p style="color: #15803d; margin-bottom: 0; font-size: 14px;">Creating a horizontal bar chart to visualize and compare the average GPA across different departments.</p>

````python
%matplotlib inline
import matplotlib.pyplot as plt

dept_gpa = df.groupby("department")["gpa"].mean().sort_values()

plt.figure(figsize=(8, 4))
plt.barh(dept_gpa.index, dept_gpa.values, color="#6366f1")
plt.xlabel("Average GPA")
plt.title("Average GPA by Department", fontweight="bold")
plt.tight_layout()
plt.show()
````

## 6. Challenge Exercise

> **Goal:** Load data → 3 transformations → 1 visualization

### Step 1: Load

<div style="background-color: #f0fdf4; padding: 15px; border-left: 5px solid #22c55e; border-radius: 5px; margin-bottom: 10px;">
    <h4 style="color: #166534; margin-top: 0;">🎯 What are we doing here?</h4>
    <p style="color: #15803d; margin-bottom: 0; font-size: 14px;">Loading our student performance dataset from a CSV file into a Pandas DataFrame and previewing the first few rows.</p>
</div>

````python
df = pd.read_csv("data/students_performance.csv")
print(f"Loaded {len(df)} students")
````

### Step 2: Transform

<div style="background-color: #f0fdf4; padding: 15px; border-left: 5px solid #22c55e; border-radius: 5px; margin-bottom: 10px;">
    <h4 style="color: #166534; margin-top: 0;">🎯 What are we doing here?</h4>
    <p style="color: #15803d; margin-bottom: 0; font-size: 14px;">Filtering for active students (high attendance) and calculating their average exam scores grouped by department.</p>
</div>

````python
# Transformation 1 — Filter: only students with attendance >= 80%
df_active = df[df["attendance_pct"] >= 80]
print(f"Active students (attendance >= 80%): {len(df_active)}")

# Transformation 2 — Group: average score by department
#computes the average exam score within each department group (using only active students, since we're working off df_active).
dept_scores = df_active.groupby("department")["exam_score"].mean().round(1)
print("\nAvg exam score by department:")
print(dept_scores)

# Transformation 3 — New column: pass/fail
df_active = df_active.copy()
df_active["status"] = df_active["exam_score"].apply(lambda x: "Pass" if x >= 70 else "Fail")
print("\nPass/Fail counts:")
print(df_active["status"].value_counts())
````

### Step 3: Visualize

<div style="background-color: #f0fdf4; padding: 15px; border-left: 5px solid #22c55e; border-radius: 5px; margin-bottom: 10px;">
    <h4 style="color: #166534; margin-top: 0;">🎯 What are we doing here?</h4>
    <p style="color: #15803d; margin-bottom: 0; font-size: 14px;">Creating a dashboard with multiple charts side-by-side to visualize both average scores and pass/fail rates at a glance.</p>
</div>

````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Chart 1: Avg score by department
dept_scores.sort_values().plot(kind="barh", ax=ax1, color="#6366f1")
ax1.set_title("Avg Exam Score by Dept", fontweight="bold")
ax1.set_xlabel("Score")

# Chart 2: Pass vs Fail
status = df_active["status"].value_counts()
ax2.pie(status, labels=status.index, autopct="%1.0f%%", colors=["#10b981","#ef4444"])
ax2.set_title("Pass vs Fail (Active Students)", fontweight="bold")

plt.tight_layout()
plt.show()

print("Insight: Students with >= 80% attendance have a high pass rate!")
````

## Part 5: GeoPandas

GeoPandas extends the datatypes used by pandas to allow spatial operations on geometric types. It makes working with geospatial data in Python easier.

In this example, we will load a built-in dataset containing the map of the world and plot it.

<div style="background-color: #f0fdf4; padding: 15px; border-left: 5px solid #22c55e; border-radius: 5px; margin-bottom: 10px;">
    <h4 style="color: #166534; margin-top: 0;">🎯 What are we doing here?</h4>
    <p style="color: #15803d; margin-bottom: 0; font-size: 14px;">Loading and displaying a world map dataset using GeoPandas to explore geospatial data handling.</p>
</div>

Run -> `poetry add geopandas geodatasets`

````python
import geopandas as gpd
import geodatasets
import matplotlib.pyplot as plt

# Load the built-in naturalearth_lowres dataset using geodatasets (new standard for GeoPandas 1.0+)
dataset_path = geodatasets.get_path('naturalearth.land')
world = gpd.read_file(dataset_path)

# Display the first few rows
display(world.head())

# Plot the world map
world.plot(figsize=(10, 6), color='lightblue', edgecolor='black')
plt.title('World Map using GeoPandas')
plt.show()
````

````python
from shapely.geometry import Point, LineString, Polygon, MultiPolygon
import matplotlib.pyplot as plt

# Create a Point
pt = Point(2, 3)

# Create a LineString
line = LineString([(0, 0), (2, 3), (4, 5)])

# Create a Polygon
poly = Polygon([(0, 0), (4, 0), (4, 4), (0, 4)])

# Create a MultiPolygon
poly1 = Polygon([(6, 0), (8, 0), (8, 2), (6, 2)])
poly2 = Polygon([(9, 0), (10, 0), (10, 1), (9, 1)])
multi_poly = MultiPolygon([poly1, poly2])

# Show differences
print("Point:", pt)
print("Type:", type(pt))
print("\nLineString:", line)
print("Type:", type(line))
print("\nPolygon:", poly)
print("Type:", type(poly))
print("\nMultiPolygon:", multi_poly)
print("Type:", type(multi_poly))

# Plot each geometry
fig, axs = plt.subplots(1, 4, figsize=(12, 4))

# Point
axs[0].scatter(*pt.xy, color='red')
axs[0].set_title("Point")
axs[0].set_xlim(-1, 5)
axs[0].set_ylim(-1, 5)

# LineString
x, y = line.xy
axs[1].plot(x, y, color='blue')
axs[1].set_title("LineString")
axs[1].set_xlim(-1, 5)
axs[1].set_ylim(-1, 6)

# Polygon
x, y = poly.exterior.xy
axs[2].fill(x, y, color='green', alpha=0.5)
axs[2].set_title("Polygon")
axs[2].set_xlim(-1, 5)
axs[2].set_ylim(-1, 5)

# MultiPolygon
for geom in multi_poly.geoms:
    x, y = geom.exterior.xy
    axs[3].fill(x, y, color='orange', alpha=0.5)
axs[3].set_title("MultiPolygon")
axs[3].set_xlim(5, 11)
axs[3].set_ylim(-1, 3)

for ax in axs:
    ax.set_aspect('equal')
    ax.axis('off')

plt.tight_layout()
plt.show()
````
